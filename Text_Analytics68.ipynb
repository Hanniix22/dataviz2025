# คำบ่นเรื่องรอนาน” โผล่บ่อยแค่ไหน และทำให้คะแนนตกจริงไหม?

````python
import re
import numpy as np
import pandas as pd
from scipy.stats import mannwhitneyu

# --- Safety cast/clean ---
df["text"] = df["text"].astype(str)
# ถ้าชื่อคอลัมน์ดาวจริงเป็น stars_review ให้แน่ใจว่าเป็นตัวเลข
df["stars_review"] = pd.to_numeric(df["stars_review"], errors="coerce")

# --- 1) สร้าง flag ด้วย regex (รองรับ EN/TH) แบบเวกเตอร์ ---
pattern = (
    r"(?i)"                                  # case-insensitive
    r"\bwait(?:ing|ed)?\b|"                  # wait, waiting, waited
    r"\blong\s*time\b|"                       # long time
    r"\bqueu(?:e|ing|e?d)\b|"                 # queue, queuing, queued
    r"\bline\b|"                              # line (ต่อคิว)
    r"\bdelay(?:ed)?\b|"                      # delay, delayed
    r"\bslow(?:\s*service)?\b|"               # slow, slow service
    r"รอนาน|ช้า|คิว|ต่อคิว"                     # ไทย
)

df["has_wait_kw"] = df["text"].str.lower().str.contains(pattern, regex=True, na=False).astype(int)

# --- 2) แจกแจงสัดส่วนดาวภายในแต่ละกลุ่ม (มี/ไม่มีคำบ่น) ---
wait_star_distribution = (
    df.groupby("has_wait_kw")["stars_review"]
      .value_counts(normalize=True)
      .unstack(fill_value=0)
      .sort_index()
)
print("Star distribution for reviews with and without 'waiting' keywords:")
display(wait_star_distribution)

# --- 3) ค่าดาวเฉลี่ยในแต่ละกลุ่ม ---
average_stars_by_wait = df.groupby("has_wait_kw")["stars_review"].mean()
print("\nAverage star rating for reviews with and without 'waiting' keywords:")
display(average_stars_by_wait)

# --- 4) สถิติทดสอบความต่าง + Effect size ---
g1 = df.loc[df["has_wait_kw"] == 1, "stars_review"].dropna().values
g0 = df.loc[df["has_wait_kw"] == 0, "stars_review"].dropna().values

if len(g1) > 0 and len(g0) > 0:
    u, p = mannwhitneyu(g1, g0, alternative="two-sided")
    print(f"\nMann–Whitney U test: U={u:.0f}, p-value={p:.3g}")

    # Cliff's delta (effect size; -1…+1, ค่า < -0.147 ≈ small negative)
    # ประมาณแบบเร็ว (สุ่มตัดขนาดถ้าข้อมูลใหญ่มาก)
    def cliffs_delta(x, y, max_pairs=2_000_000):
        x = np.asarray(x); y = np.asarray(y)
        nx, ny = len(x), len(y)
        # sample ถ้าใหญ่
        if nx * ny > max_pairs:
            rng = np.random.default_rng(42)
            x = rng.choice(x, size=min(nx, int(np.sqrt(max_pairs))), replace=False)
            y = rng.choice(y, size=min(ny, int(np.sqrt(max_pairs))), replace=False)
            nx, ny = len(x), len(y)
        # นับคู่
        gt = sum((xi > y).sum() for xi in x)
        lt = sum((xi < y).sum() for xi in x)
        return (gt - lt) / (nx * ny)

    delta = cliffs_delta(g1, g0)
    print(f"Cliff’s delta = {delta:.3f} (negative = กลุ่มที่มีคำบ่นได้ดาวต่ำกว่า)")
else:
    print("\nNot enough data for statistical testing.")

# --- 5) กราฟสั้น ๆ (ถ้าต้องการ) ---
import matplotlib.pyplot as plt

avg = average_stars_by_wait.rename(index={0:"No-wait keywords", 1:"Has wait keywords"})
ax = avg.plot(kind="bar")
ax.set_ylabel("Average stars")
ax.set_title("Average rating by waiting-complaint flag")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
````

#ความยาวข้อความสัมพันธ์กับคะแนนไหม?

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import spearmanr

TEXT, STARS = "text", "stars_review" # Corrected column name for stars
df["len_tokens"] = df[TEXT].astype(str).str.split().apply(len)

rho, p = spearmanr(df["len_tokens"], df[STARS])
print(f"Spearman r={rho:.3f}, p={p:.3g}")

p99 = df["len_tokens"].quantile(0.99)
df_clip = df.assign(len_clip=np.minimum(df["len_tokens"], p99))
df_clip.boxplot(column="len_clip", by="stars_review")
plt.title("Length (clipped at P99) vs Stars"); plt.suptitle("")
plt.ylabel("Token length (<=P99)")
plt.show()
````

(Boxplot แบบตัดหางที่ P99):

แนวโน้มชัด: ยิ่งดาวสูง ความยาวข้อความยิ่งสั้นลง
มัธยฐานโดยคร่าว ๆ ไล่จาก ~90–100 tokens (1–2★) → ~70–80 (3–4★) → ~45–55 (5★)

กลุ่มดาวต่ำ กระจายกว้างกว่า (IQR ใหญ่) และยังมีหางยาว (แม้จะ clip ที่ P99 แล้ว) ⇒ คนให้ดาวต่ำมักเขียนยาวและเล่าเยอะ
