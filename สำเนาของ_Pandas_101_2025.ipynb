#HW 4 (group) ยอดรวมทั้ง 5 เดือน เมืองไหนมีจำนวนผู้ถูกเรียกค้นที่เป็นคน Asia น้อยที่สุด และ มากที่สุด

````
import os
import pandas as pd
# โดยเราจะกำหนดที่อยู่ของ main folder เเล้ว list ไฟล์ที่เราต้องการตรวจสอบ
base_path = "/content/drive/MyDrive/Data"
months = ['2025-01', '2025-02', '2025-03', '2025-04', '2025-05']
#และสร้าง dictionary เปล่า ไว้เก็บผลลัพธ์
asian_counts = {}

for month in months:
    month_path = os.path.join(base_path, month)

    if os.path.exists(month_path):
        for filename in os.listdir(month_path):
            if filename.endswith('.csv') and 'stop-and-search' in filename:
                file_path = os.path.join(month_path, filename)

                city = filename.replace(f"{month}-", "").replace('-stop-and-search.csv', '').replace('-', ' ').title()

                try:
                    df = pd.read_csv(file_path)

                    if 'Self-defined ethnicity' in df.columns:
                        asian_df = df[df['Self-defined ethnicity']
                                        .astype(str)
                                        .str.lower()
                                        .str.contains('asian')]

                        asian_counts[city] = asian_counts.get(city, 0) + len(asian_df)

                except Exception as e:
                    print(f"❌ อ่านไฟล์ {filename} ไม่สำเร็จ: {e}")
    else:
        print(f"⚠️ ไม่มีโฟลเดอร์: {month_path}")

# 🔽 แสดงผลลัพธ์
result_df = pd.DataFrame.from_dict(asian_counts, orient='index', columns=['Asian Count'])

if not result_df.empty:
    result_df = result_df.sort_values('Asian Count', ascending=False)

    print("✅ สรุปจำนวนผู้ถูกค้นที่เป็นคน Asian (Self-defined) ต่อเมือง (รวมทุกเดือน):")
    print(result_df)

    most = result_df.idxmax().values[0]
    least = result_df.idxmin().values[0]

    print(f"\n📌 มากที่สุด: {most} ({result_df.loc[most, 'Asian Count']} คน)")
    print(f"📌 น้อยที่สุด: {least} ({result_df.loc[least, 'Asian Count']} คน)")
else:
    print("⚠️ ไม่พบข้อมูลคน Asian (Self-defined) ในไฟล์ใด ๆ เลย")
````

````
import pandas as pd #pandas สำหรับจัดการตารางข้อมูล
import os #os สำหรับจัดการไฟล์/โฟลเดอร์

#ดึงข้อมูลจากทุกไฟล์ .csv ที่เกี่ยวกับ stop-and-search จากทุกโฟลเดอร์ย่อย (แต่ละเดือน) แล้วรวมกันเป็น list ของ DataFrame พร้อมเพิ่มชื่อเมืองในคอลัมน์ "Police Force"
def load_all_data(folder_path):
    combined_data = []
    month_dirs = os.listdir(folder_path)

    #วนลูปในโฟลเดอร์รายเดือน
    for month in month_dirs:
        full_month_path = os.path.join(folder_path, month)
        filenames = os.listdir(full_month_path)

        #ตรวจหาไฟล์ .csv ที่มีคำว่า 'stop-and-search'
        for name in filenames:

            # เลือกเฉพาะไฟล์ที่เป็น .csv และมีคำว่า 'stop-and-search' ในชื่อ
            if name.endswith(".csv") and 'stop-and-search' in name:

                #อ่านข้อมูลจากไฟล์ และเพิ่มชื่อเมือง
                file_path = os.path.join(full_month_path, name)
                df = pd.read_csv(file_path)

                # ตัดชื่อเมืองออกมาจากชื่อไฟล์
                city_name = name.replace('-stop-and-search.csv', '').split('-', 2)[-1]
                df['Police Force'] = city_name  # เพิ่มชื่อเมืองลงในคอลัมน์ใหม่
                combined_data.append(df)
    #ส่งออกเป็น list ของ DataFrame
    return combined_data

#นับจำนวนการถูกตรวจค้นของบุคคลที่ระบุว่าเป็นเชื้อสายเอเชีย
def count_ethnicity_asian(data_list):
    summary = {}

    for table in data_list:
        for _, row in table.iterrows():
            ethnicity = str(row.get('Self-defined ethnicity', ''))
            if 'Asian' in ethnicity:
                city = row['Police Force']
                summary[city] = summary.get(city, 0) + 1

    return summary

#-------- จุดเริ่มต้นของโปรแกรม --------
root_path = '/content/drive/MyDrive/Data/'

#โหลดข้อมูลทุกเดือน
all_data = load_all_data(root_path)

#ประมวลผลจำนวนคนเชื้อสายเอเชียที่ถูกตรวจค้น
asian_summary = count_ethnicity_asian(all_data)

#หาค่ามากสุดและน้อยสุด
top_city = max(asian_summary, key=asian_summary.get)
bottom_city = min(asian_summary, key=asian_summary.get)

#แสดงผลลัพธ์
print(">>> จำนวนการตรวจค้นบุคคลเชื้อสายเอเชียในแต่ละเมือง (รวมหลายเดือน):\n")
for city in sorted(asian_summary, key=asian_summary.get, reverse=True):
    print(f"{city}: {asian_summary[city]}")

print("\n=== สรุปข้อมูล ===")
print(f"เมืองที่มีการเรียกค้นคนเชื้อสายเอเชียมากที่สุด: {top_city} ({asian_summary[top_city]})")
print(f"เมืองที่มีการเรียกค้นคนเชื้อสายเอเชียน้อยที่สุด: {bottom_city} ({asian_summary[bottom_city]})")
````

#HW5 ยอดรวมเมืองทั้งหมด เดือนไหนมีจำนวนผู้ถูกเรียกค้นที่เป็นคน Asia น้อยที่สุด และ มากที่สุด

````
import pandas as pd #pandas สำหรับจัดการตารางข้อมูล
import os #os สำหรับจัดการไฟล์/โฟลเดอร์

# ดึงข้อมูลจากทุกไฟล์ .csv ที่เกี่ยวกับ stop-and-search จากทุกโฟลเดอร์ย่อย (แต่ละเดือน) แล้วรวมกันเป็น list ของ DataFrame พร้อมเพิ่มชื่อเดือนในคอลัมน์ "Month"
def load_all_data(folder_path):
    combined_data = []
    month_dirs = os.listdir(folder_path)

    #วนลูปในโฟลเดอร์รายเดือน
    for month in month_dirs:
        full_month_path = os.path.join(folder_path, month)
        filenames = os.listdir(full_month_path)

        #ตรวจหาไฟล์ .csv ที่มีคำว่า 'stop-and-search'
        for name in filenames:

            # เลือกเฉพาะไฟล์ที่เป็น .csv และมีคำว่า 'stop-and-search' ในชื่อ
            if name.endswith(".csv") and 'stop-and-search' in name:

                #อ่านข้อมูลจากไฟล์ และเพิ่มชื่อเมือง
                file_path = os.path.join(full_month_path, name)
                df = pd.read_csv(file_path)

                # เพิ่มชื่อเดือนลงในคอลัมน์ใหม่
                df['Month'] = month
                combined_data.append(df)
    #ส่งออกเป็น list ของ DataFrame
    return combined_data

#นับจำนวนการถูกตรวจค้นของบุคคลที่ระบุว่าเป็นเชื้อสายเอเชีย โดยรวมตามเดือน
def count_ethnicity_asian_by_month(data_list):
    summary = {}

    for table in data_list:
        month = table['Month'].iloc[0] # Get the month from the added column
        for _, row in table.iterrows():
            ethnicity = str(row.get('Self-defined ethnicity', ''))
            if 'Asian' in ethnicity:
                summary[month] = summary.get(month, 0) + 1

    return summary

#-------- จุดเริ่มต้นของโปรแกรม --------
root_path = '/content/drive/MyDrive/Data/'

#โหลดข้อมูลทุกเดือน
all_data = load_all_data(root_path)

#ประมวลผลจำนวนคนเชื้อสายเอเชียที่ถูกตรวจค้นตามเดือน
asian_summary_by_month = count_ethnicity_asian_by_month(all_data)

#หาค่ามากสุดและน้อยสุด
top_month = max(asian_summary_by_month, key=asian_summary_by_month.get)
bottom_month = min(asian_summary_by_month, key=asian_summary_by_month.get)

#แสดงผลลัพธ์
print(">>> จำนวนการตรวจค้นบุคคลเชื้อสายเอเชียในแต่ละเดือน (รวมทุกเมือง):\n")
for month in sorted(asian_summary_by_month):
    print(f"{month}: {asian_summary_by_month[month]}")

print("\n=== สรุปข้อมูล ===")
print(f"เดือนที่มีการเรียกค้นคนเชื้อสายเอเชียมากที่สุด: {top_month} ({asian_summary_by_month[top_month]})")
print(f"เดือนที่มีการเรียกค้นคนเชื้อสายเอเชียน้อยที่สุด: {bottom_month} ({asian_summary_by_month[bottom_month]})")
````

````
# prompt: มีคำดังกล่าวใน นิทานชีวิต หรือ นิทานประจำถิ่นมากกว่ากัน บอกจำนวนด้วย

import pandas
import pandas as pd
import os
from google.colab import drive
from bs4 import BeautifulSoup

# ... (preceding code)

count_life = 0
count_local = 0
for index, row in this_data.iterrows():
  if ('แม่' in row['title'] or 'พ่อ' in row['title']) and row['type'] == 'นิทานชีวิต':
    count_life += 1
  elif ('แม่' in row['title'] or 'พ่อ' in row['title']) and row['type'] == 'นิทานประจำถิ่น':
    count_local += 1

if count_life > count_local:
  print(f"There are more stories with 'แม่' or 'พ่อ' in the title in นิทานชีวิต ({count_life} stories) than in นิทานประจำถิ่น ({count_local} stories).")
elif count_local > count_life:
  print(f"There are more stories with 'แม่' or 'พ่อ' in the title in นิทานประจำถิ่น ({count_local} stories) than in นิทานชีวิต ({count_life} stories).")
else:
  print(f"There are an equal number of stories with 'แม่' or 'พ่อ' in the title in both นิทานชีวิต and นิทานประจำถิ่น ({count_life} stories each).")
````

````
# prompt: คนแต่งคนไหน ชอบแต่งนิทานที่มีชื่อเรื่อง ว่า แม่ หรือ พ่อ มากที่สุด (พร้อมทั้งอธิบาย code แต่ละบรรทัดด้วย)

# สร้าง dictionary เพื่อเก็บจำนวนนิทานที่มีคำว่า 'แม่' หรือ 'พ่อ' ในชื่อเรื่องสำหรับแต่ละคนแต่ง
creator_counts = {}
for index, row in this_data.iterrows():
  if 'แม่' in row['title'] or 'พ่อ' in row['title']:
    creator = row['creator']
    if creator not in creator_counts:
      creator_counts[creator] = 0
    creator_counts[creator] += 1

# หาคนแต่งที่มีจำนวนนิทานมากที่สุด
max_count = 0
most_prolific_creator = None
for creator, count in creator_counts.items():
  if count > max_count:
    max_count = count
    most_prolific_creator = creator

print(f"The creator who wrote the most stories with 'แม่' or 'พ่อ' in the title is {most_prolific_creator} with {max_count} stories.")
````
